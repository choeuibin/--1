```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
```

```python
df = pd.read_excel("/content/drive/MyDrive/프로젝트2/new_member_marketing_1.xlsx")
df
```

```python
df1 = df[df["출생년도"] != "-"]
df1 = df1[df1["관심자격1"] != "-"]
df1 = df1[df1["관심자격2"] != "-"]
df1 = df1[df1["관심자격3"] != "-"]
```

```python
value_list = df1["출생년도"].unique()
print(value_list)
count = pd.DataFrame(df1["출생년도"].value_counts())
count.head(10)
```

```python
value_list = df1["관심자격1"].unique()
print(value_list)
count = pd.DataFrame(df1["관심자격1"].value_counts())
count.head(10)
```

```python
value_list = df1["관심자격2"].unique()
print(value_list)
count = pd.DataFrame(df1["관심자격2"].value_counts())
count.head(10)
```

```python
value_list = df1["관심자격3"].unique()
print(value_list)
count = pd.DataFrame(df1["관심자격3"].value_counts())
count.head(10)
```

```python
df_choice = pd.DataFrame(df1.groupby("출생년도")[["관심자격1","관심자격2", "관심자격3"]].value_counts())
```
