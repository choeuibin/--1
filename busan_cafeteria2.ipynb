```python
#주변 맛집 리스트에 넣어주기
food_list = []
for i in nearby:
  title = i["title"]
  category = i["category"]
  roadAddress = i['roadAddress']
  
  data = {
      'title' : title.replace('<b>', '').replace('</b>', ''),
      'category' : category.replace('<b>', '').replace('</b>', ''),
      'roadAddress' : roadAddress
  }

  food_list.append(data)
```

```python
import json
import urllib
from urllib.request import Request, urlopen
from urllib.parse import quote

start = '부산 해운대구 APEC로 55'
goal = '부산광역시 해운대구 우동'
waypoint = '부산광역시 해운대구 해운대로 522'

def get_location(loc) :
    client_id = 'p5ct5bab1q'
    client_secret = 'kOStvHtf2OXLpdK1WOGJwaWEVPu7t3tk8Uw8j3EX'
    url = f"https://naveropenapi.apigw.ntruss.com/map-geocode/v2/geocode?query=" \
    			+ urllib.parse.quote(loc)
    
    request = urllib.request.Request(url)
    request.add_header('X-NCP-APIGW-API-KEY-ID', client_id)
    request.add_header('X-NCP-APIGW-API-KEY', client_secret)
    
    response = urlopen(request)
    res = response.getcode()
    
    if (res == 200) : 
        response_body = response.read().decode('utf-8')
        response_body = json.loads(response_body)
        print(response_body)
  
        if response_body['meta']['totalCount'] == 1 : 
            lat = response_body['addresses'][0]['y']
            lon = response_body['addresses'][0]['x']
            return (lon, lat)
        else :
            print('location not exist')
        
    else :
        print('ERROR')
        
#  함수 적용
start = get_location(start)
goal = get_location(goal)
waypoint = get_location(waypoint)
```

```python
#waypoint list 만들기
waypoint_list = []
for i in food_list:
  roadAddress = i['roadAddress']
  waypoint = get_location(roadAddress)
  #waypoint = list(waypoint)
  waypoint_list.append(waypoint)
```

```python
# *-- Directions 5 활용 코드 --*
option = 'traavoidcaronly'
# option : 탐색옵션 [최대 3개, traoptimal(기본 옵션) 
# / trafast, tracomfort, traavoidtoll, traavoidcaronly]

def get_optimal_route(start, goal, waypoint, option) :
    client_id = 'p5ct5bab1q'
    client_secret = 'kOStvHtf2OXLpdK1WOGJwaWEVPu7t3tk8Uw8j3EX' 
    # start=/goal=/(waypoint=)/(option=) 순으로 request parameter 지정
    start_url = f"https://naveropenapi.apigw.ntruss.com/map-direction-15/v1/driving? \
    start={start[0]},{start[1]}&goal={goal[0]},{goal[1]}\
    &option={option}&waypoints={waypoint[0]},{waypoint[1]}"
    url = start_url.replace(" ","")
    print(url)
    request = urllib.request.Request(url)
    #print(request)
    request.add_header('X-NCP-APIGW-API-KEY-ID', client_id)
    request.add_header('X-NCP-APIGW-API-KEY', client_secret)
    
    response = urllib.request.urlopen(request)
    #print(response)
    res = response.getcode()
    #print(res)
    
    if (res == 200) :
        response_body = response.read().decode('utf-8')
        print(response_body)
        return json.loads(response_body)
            
    else :
        print('ERROR')
```

```python
#waypoint 넣고 시간 조회 
#duration : 1/1000 초
time_list = []

for i in waypoint_list:
  results = get_optimal_route(start, goal, waypoint=i, option=option)
  print(results)
  total_duration = results['route']['traavoidcaronly'][0]['summary']['duration']/1000/60
  print(total_duration)

  time_list.append(total_duration)
```
